```julia
using CSV
using JuMP
#using CPLEX
using GLPK
using Pkg
using Ipopt

DATA = CSV.read("pred_cord.csv")

COOR = convert(Matrix, DATA[["coord_X","coord_Y"]])
PROD = convert(Matrix,DATA[["Pro"]])/10000
PROD = convert.(Int64,PROD)

costos = [250,300,350]

n = size(COOR)[1]
costos[PROD[1]]
#PREGUNTA 1:
modelo = Model()
set_optimizer(modelo, Ipopt.Optimizer)

@variable(modelo,x)
@variable(modelo,y)
@variable(modelo, c>=0)

@NLconstraint(modelo,sum(sqrt((COOR[i,1]-x)^2+(COOR[i,2]-y)^2)*1.1*costos[PROD[i]]*2 for i=1:n)<=c)

@NLobjective(modelo,Min,c)
optimize!(modelo)

print("-Las coordenadas que minimizan el costo, son X e Y iguales a ",round(value(x),digits = 4)," y ",round(value(y),digits=4)," respectivamente.\n-El valor de la funcion objetivo es US ",round(value(c),digits=4))
```

```julia
n = convert.(Int64,n)

#function distReal(a,b,c)
#    return sqrt((COOR[a,1]-b)^2+(COOR[a,2]-c)^2)*1.1
#end
function menor(d,e,f)
    if(sqrt((COOR[d,1]-e)^2+(COOR[d,2]-f)^2)*1.1 < sqrt((COOR[d,1]-value(x))^2+(COOR[d,2]-value(y))^2)*1.1)
        return sqrt((COOR[d,1]-e)^2+(COOR[d,2]-f)^2)*1.1
    else
        return sqrt((COOR[d,1]-value(x))^2+(COOR[d,2]-value(y))^2)*1.1
    end
end

modelo2 = Model()

JuMP.register(modelo2, :menor,3,menor,autodiff=true)
#JuMP.register(modelo2, :distReal,3,distReal,autodiff=true)

set_optimizer(modelo2, Ipopt.Optimizer)

@variable(modelo2,x2)
@variable(modelo2,y2)
@variable(modelo2, c2>=0)

@NLconstraint(modelo2,sum(menor(i,x2,y2)*costos[PROD[i]]*2 for i=1:n)<=c2)

@NLobjective(modelo2,Min,c2)
optimize!(modelo2)

print("-Las coordenadas que minimizan el costo, son X e Y iguales a ",round(value(x2),digits = 4)," y ",round(value(y2),digits=4)," respectivamente.\n-El valor de la funcion objetivo es US ",round(value(c2),digits=4))
```
